```python
import pandas as pd

df = pd.read_csv('titanic survival.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df = df[['Survived', 'Age', 'Sex', 'Pclass']]
df = pd.get_dummies(df, columns=['Sex','Pclass'])
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('Survived',axis=1)
y= df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

# Build and Train a model

Our classifier will use logistic regression. Logistic regression not only makes predictions, it yield probabilities as well.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

Get an accuracy score.

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, x, y, cv=5).mean()
```

Use a confusion matrix to see how the model performed during testing

```python
from sklearn.metrics import confusion_matrix

y_predicted = model.predict(x_test)
confusion_matrix(y_test, y_predicted)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predicted))
```

```python
female = [[80, 1, 0 , 1, 0, 0]]
model.predict(female)[0]
```

```python
probability = model.predict_proba(female)[0][1]
print(f'Probability of survival: {probability: .1%}')
```
